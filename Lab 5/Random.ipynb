```python
"""
Problem 1:  
Write a function that simulates rolling two dice and returns their sum. Use the function to simulate 1000 rolls and return a dictionary showing how many times each sum (2-12) appeared. 
Output: 2: 28, 3: 55, 4: 82, ..., 12: 27} 
"""

import random

def roll_two_dice():
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    return die1 + die2

results = {sum_: 0 for sum_ in range(2, 13)}

for _ in range(1000):
    roll_sum = roll_two_dice()
    results[roll_sum] += 1
print("Sum of two dice rolls:", results)
```

```python
"""
Problem 2:  
Write a function that generates a random password of given length containing uppercase, lowercase, digits, and special characters. 
Input: 12 
Output: "aB3$xY9#mK2@" 
"""
import random
import string

def generate_password(length):
    characters = (
        string.ascii_lowercase +
        string.ascii_uppercase +
        string.digits +
        string.punctuation
    )
    
    password = ""
    for _ in range(length):
        password += random.choice(characters)
    
    return password

password_length = 12
generated_password = generate_password(password_length)
print("Generated Password:", generated_password)
```

```python
"""
Problem 3:  
Write a function that takes a list of students and randomly divides them into groups of a specified size. If students don't divide evenly, the last group can be smaller. 
Input: ["Ahmed", "Sara", "Omar", "Ali", "Laila"], group_size=2 
Output: [["Ahmed", "Sara"], ["Omar", "Ali"], ["Laila"]] 
"""
import random

def divide_into_groups(students, group_size):
    random.shuffle(students)
    
    groups = []
    for i in range(0, len(students), group_size):
        group = students[i:i + group_size]
        groups.append(group)
    
    return groups
students_list = ["Ahmed", "Sara", "Omar", "Ali", "Laila"]
group_size = 2  
groups = divide_into_groups(students_list, group_size)
print("Groups:", groups)
```

```python
"""
Problem 4:  
Write a function that simulates a lottery where you pick 6 numbers from 1-49. Generate a random winning combination and check how many numbers the user matched. 
User picks: [5, 12, 23, 34, 45, 49] 
Winning numbers: [12, 23, 30, 34, 42, 49] 
Output: "You matched 4 numbers!" 
"""
import random

def lottery(user_picks):
    winning_numbers = random.sample(range(1, 50), 6)
    
    matched_numbers = set(user_picks) & set(winning_numbers)
    matched_count = len(matched_numbers)
    
    return winning_numbers, f"You matched {matched_count} numbers!"
user_picks = [5, 12, 23, 34, 45, 49]
winning_numbers, result = lottery(user_picks)
print("Winning numbers:", winning_numbers)
print(result)

```

```python
"""
Problem 5:  
Write a function that generates random quiz questions. Given a list of questions with answers, select 5 random questions and return them without duplicates. 
Input: 20 questions 
Output: 5 randomly selected questions 
"""

import random

def generate_quiz(questions):
    return random.sample(questions, 5)


questions = [
    "What is 2 + 2?",
    "Capital of Egypt?",
    "Largest planet in the solar system?",
    "Who invented Python?",
    "Fastest land animal?",
    "Square root of 16?",
    "Opposite of hot?",
    "Water chemical formula?",
    "Sun is a?",
    "How many days in a week?",
    "Capital of France?",
    "Largest ocean?",
    "First month of the year?",
    "HTML stands for?",
    "Speed of light?",
    "Tallest animal?",
    "Hardest natural substance?",
    "Smallest planet?",
    "Gas we breathe?",
    "Binary of 2?"
]

selected_questions = generate_quiz(questions)

print("Random Quiz Questions:")
for q in selected_questions:
    print("-", q)
```
