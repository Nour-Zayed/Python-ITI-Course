#### Classes and Objects 

```python
"""
Problem 1: 
Write a class Book with attributes: title, author, and price. Create a method that applies a discount percentage to the price and returns the new price. 
Book("Python Basics", "John Doe", 100) 
apply_discount(20) → Output: 80.0 
"""
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def apply_discount(self, discount_percentage):
        discount_amount = (discount_percentage / 100) * self.price
        new_price = self.price - discount_amount
        return new_price


book = Book("Python Basics", "John Doe", 100)
print(book.apply_discount(20))




```

```python
"""
Problem 2: 
Write a class BankAccount that initializes with an account holder's name and balance. Create methods to deposit, withdraw, and check balance. Withdrawal should only work if sufficient balance exists. 
Account("Ahmed", 1000) 
withdraw(200) → Balance: 800 
withdraw(1000) → "Insufficient funds" 
"""

class BankAccount:
    def __init__(self, Name_holder, balance):
        self.Name_holder = Name_holder
        self.balance = balance
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited: {amount}. New Balance: {self.balance}")
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrew: {amount}. New Balance: {self.balance}")
    def check_balance(self):
        print(f"Balance: {self.balance}")
account = BankAccount("Ahmed", 1000)
account.withdraw(200)  # Balance ---> 800
account.withdraw(1000)  



```

```python
"""
Problem 3: 
Write a class Rectangle with attributes width and height. Create methods to calculate area, perimeter, and check if it's a square. 
Rectangle(5, 5) 
area() → 25 
is_square() → True 
"""
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
    def Perimeter(self):
        return 2 * (self.width + self.height)
    def is_square(self):
        return self.width == self.height
rect = Rectangle(5, 5)
print(rect.area())       
print(rect.is_square())        
```

```python
"""
Problem 4: 
Write a class Temperature that initializes with Celsius. Create methods to convert to Fahrenheit and Kelvin. 
Formulas: F = C × 9/5 + 32, K = C + 273.15 
Temperature(25) 
to_fahrenheit() → 77.0 
to_kelvin() → 298.15
 """
class Temperature:
    def __init__(self, Celsius):
        self.celsius = Celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32    

    def to_kelvin(self):
        return self.celsius + 273.15
    
temp = Temperature(25)
print(temp.to_fahrenheit())
print(temp.to_kelvin())
```

#### Inheritance 

```python
"""
Problem 1: 
Create a class Person with name and age. Create a child class Employee that adds salary and job_title. Use super() to initialize parent attributes. Create a method to give a raise (increase salary by a percentage). 
Employee("Ahmed", 30, 5000, "Engineer") 
give_raise(10) → New salary: 5500 
"""
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
class Employee(Person):
    def __init__(self,name, age, Salary,job_title):
        super().__init__(name,age)
        self.salary = Salary
        self.job_title = job_title
    def Salary_raise(self, percentage):
        raise_amount = (percentage / 100) * self.salary
        self.salary += raise_amount
        print(f"New salary: {self.salary}")

employee = Employee("Ahmed", 30, 5000, "Engineer")
employee.Salary_raise(10)  
            
```

```python
"""
Problem 2: 
Create a multilevel inheritance: Person → Student → GraduateStudent. Each level should add new attributes (Person: name, Student: +grades, GraduateStudent: +thesis_topic). Use super() appropriately. 
GraduateStudent("Ali", [85, 90], "AI Research") 
Display all information from all levels 
"""
class Person:
    def __init__(self, name):
        self.name = name
class Student(Person):
    def __init__(self,name,grades):
            super().__init__(name)
            self.grades = grades
class GraduateStudent(Student):
    def __init__(self,name,grades,thesis_topic):
            super().__init__(name,grades)
            self.thesis_topic = thesis_topic
    def display_info(self):
            print(f"Name: {self.name}")
            print(f"Grades: {self.grades}")
            print(f"Thesis Topic: {self.thesis_topic}")

grad_student = GraduateStudent("Ali", [85, 90], "AI Research")
grad_student.display_info()            

```

```python
"""
Problem 3: 
Create multiple inheritance: Teacher (has subject) and Researcher (has research_area). Create ProfessorTeacher that inherits from both and can display information from both parent classes. 
Professor("Ahmed", "Math", "Number Theory") 
Display: teaches Math, researches Number Theory 
"""
class Teacher:
    def __init__(self,name,subject):
        self.name = name
        self.subject = subject
class Researcher:
    def __init__(self,name,research_area):
        self.name = name
        self.research_area = research_area
class ProfessorTeacher(Teacher,Researcher):
    def __init__(self,name,subject,research_area):
        Teacher.__init__(self,name,subject)
        Researcher.__init__(self,name,research_area)
    def display_info(self):
        print(f"{self.name} teaches {self.subject}, researches {self.research_area}")

professor = ProfessorTeacher("Ahmed", "Math", "Number Theory")
professor.display_info()
```

```python
"""
Problem 1:
 Create a function calculate_area(shape) that takes any shape object and calls its area() method. Create classes Triangle, Circle, and Rectangle each with their own area calculation. 
 calculate_area(Circle(5)) → 78.5 calculate_area(Rectangle(4, 5)) → 20
"""

import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


def calculate_area(shape):
    return shape.area()


print(calculate_area(Circle(5)))       
print(calculate_area(Rectangle(4, 5))) 
```

#### Encapsulation 

```python
"""
Problem 1: 
Create a class PasswordManager with a private attribute password. Create a method to set password (must be at least 8 characters ) and a method to verify password (check if input matches). 
set_password("abc") → "Password too short" 
set_password("secure123") → "Password set" 
verify("secure123") → True
 """
class PasswordManager:
    def __init__(self):
        self._password = ""
    def set_password(self, password):
        if len(password) < 8:
            return "Password too short"
        else:
            self._password = password
            return "Password set"
        
    def verify(self, password):
        return self._password == password
pm = PasswordManager()
print(pm.set_password("abc"))    
print(pm.set_password("secure123"))  
print(pm.verify("secure123"))        

```

```python
"""
Problem 2: 
Create a class Counter with a private attribute count starting at 0. Create methods to increment, decrement, and reset the counter. The count should never go below 0. 
counter.increment() → Count: 1 
counter.decrement() → Count: 0 
counter.decrement() → Count: 0 (doesn't go negative) 
"""

class Counter:
    def __init__(self):
        self.__count = 0
    def increment(self):
        self.__count += 1
        print(f"Count: {self.__count}")
    def decrement(self):
        if self.__count > 0:
            self.__count -= 1
        print(f"Count: {self.__count}")
    def reset(self):
        self.__count = 0
        print(f"Count: {self.__count}")

counter = Counter()
counter.increment()
counter.decrement()
counter.decrement() 
counter.reset()
```

#### Abstraction 

```python
"""
Problem 1: 
Create an abstract class Shape with an abstract method area() and a concrete method describe() that returns "This is a shape". Create classes Circle and Square that implement area but inherit describe. 
Circle(5).area() → 78.5 
Circle(5).describe() → "This is a shape" 
"""

from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    def describe(self):
        return "This is a shape"
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius
    

    class Square(Shape):
        def __init__(self, side):
             self.side = side

        def area(self):
            return self.side * self.side
        

circle = Circle(5)
print(circle.area())
print(circle.describe())        
    

```

```python
"""
Problem 2: 
Create an abstract class Employee with an abstract property salary and an abstract method calculate_bonus(). Create concrete classes Manager (bonus = 20% of salary) and Developer (bonus = 10% of salary). 
Manager(salary=10000).calculate_bonus() → 2000 
Developer(salary=8000).calculate_bonus() → 800 
"""

from abc import ABC, abstractmethod

class Employee(ABC):

    @property
    @abstractmethod
    def salary(self):
        pass

    @abstractmethod
    def calculate_bonus(self):
        pass
class Manager(Employee):
    def __init__(self, salary):
        self._salary = salary

    @property
    def salary(self):
        return self._salary

    def calculate_bonus(self):
        return 0.2 * self.salary

class Developer(Employee):
    def __init__(self, salary):
        self._salary = salary

    @property
    def salary(self):
        return self._salary

    def calculate_bonus(self):
        return 0.1 * self.salary    

manager = Manager(10000)
developer = Developer(8000)
print(manager.calculate_bonus())
print(developer.calculate_bonus())    
```

```python
"""
Problem 3: 
Create an abstract class FileHandler with abstract methods read() and write(data). Create concrete classes TextFileHandler and JSONFileHandler that implement these methods appropriately. 
TextFileHandler().write("Hello") → "Writing text: Hello" 
JSONFileHandler().write({"key": "value"}) → "Writing JSON: {...}" 
"""

from abc import ABC, abstractmethod

class FileHandler(ABC):

    @abstractmethod
    def read(self):
        pass

    @abstractmethod
    def write(self, data):
        pass


class TextFileHandler(FileHandler):

    def read(self):
        return "Reading text file"

    def write(self, data):
        return f"Writing text: {data}"


class JSONFileHandler(FileHandler):

    def read(self):
        return "Reading JSON file"

    def write(self, data):
        return f"Writing JSON: {data}"

text = TextFileHandler()
json_file = JSONFileHandler()

print(text.write("Hello"))
print(json_file.write({"key": "value"}))
```
