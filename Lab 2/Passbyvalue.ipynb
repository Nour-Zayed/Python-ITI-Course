```python
"""""
Problem 1:  
Write a program to test how different types behave when passed to a function. 
Create variables name, age, salary.  
Write a function that tries to change the value of a variable inside it then print the results before and after. 
"""""

name = "Nour"
age = 22
salary = 5000.0

def change_values(name, age, salary):
    print("\nInside function (before change):")
    print(name, age, salary)

    name = "Sara"
    age = 30
    salary = 8000.0

    print("Inside function (after change):")
    print(name, age, salary)

print("Before function call:")
print(name, age, salary)

change_values(name, age, salary)

print("\nAfter function call:")
print(name, age, salary)

```

```python
"""""
problem 2:  
Write a function that takes a list of numbers and replaces all odd numbers with the value print the list before and after replace . 
"""""
def replace_odd_num(numbers):
    print("Before replace:", numbers)

    for i in range(len(numbers)):
        if numbers[i] % 2 != 0:   
            numbers[i] = 0    

    print("After replace:", numbers)

# List
nums = [1, 2, 3, 4, 5, 6, 7]

replace_odd_num(nums)

print("\nList outside function:", nums)

```

```python
"""""
Problem 3:  
Write a function that takes a string and tries to change it to "Modified" inside the function. Print the string before and after calling the function to see what happens. 
Question: Did the original string change? Why or why not? 
"""""
# String Immutable
my_string = "Original"
print("before",my_string)
def modify_string(s):
    s = "Modified"
    print("inside",s)
modify_string(my_string)
print("after",my_string)    
```

```python
"""""
Problem 4:  
Write two functions: 
modify_list(lst) - adds the number 100 to the end of the list 
modify_number(num) - adds 100 to the number 
"""""
def modify_list(lst):
    lst.append(100)
    print("Inside modify_list:", lst)

def modify_number(num):
    num += 100
    print("Inside modify_number:", num)

my_list = [1, 2, 3]
my_number = 50

print("Before modify_list:", my_list)
modify_list(my_list)
print("After modify_list:", my_list)

print("\nBefore modify_number:", my_number)
modify_number(my_number)
print("After modify_number:", my_number)


```
