```python
"""""
Problem 4:  
write a function that takes a number as an argument and if the number divisible by 3 
return "Fizz" and if it is divisible by 5 return "buzz" and if is is divisible by both return "FizzBuzz"
 """""
def fizz_buzz(num):
    if num % 3 == 0 and num % 5 == 0:
        return "FizzBuzz"
    elif num % 3 == 0:
        return "Fizz"
    elif num % 5 == 0:
        return "Buzz"
    else:
        return str(num)
Number = int(input("Enter a number: "))   
print(fizz_buzz(Number)) 

```

```python
"""""
Problem 5:  
Write a function which has an input of a string from user then it will return the same string reversed using two ways. 
"""""
def reversed_string(s):
    reversed_slice = s[::-1]
    reversed_loop = ''

    for char in s:
        reversed_loop = char + reversed_loop

    return reversed_slice, reversed_loop


input_string = input("Enter a string: ")
reversed_by_slice, reversed_by_loop = reversed_string(input_string)

print("Reversed by slicing:", reversed_by_slice)
print("Reversed by loop:", reversed_by_loop)

    

```

```python
"""""
Problem 6:  
Write a Python program that solves the following tasks without using loops: 
Calculate the result of a mathematical operation that multiplies a number by all smaller positive numbers until it reaches 1. 
Display numbers starting from 1 up to a given number. 
Find a value in a famous numerical sequence where each number is the sum of the two previous ones. 
"""""

def factorial(n):
    if n == 1:
        return 1
    return n* factorial(n-1)
# Recursion
def print_numbers(n, current=1):
    if current >n:
        return
    print(current)
    print_numbers(n, current +1)

def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

    
print(factorial(5))
print_numbers(5)
print(fibonacci(6))
  


```

```python
"""""
Problem 7:  
Write a program that continuously accepts integer values from the user and determines the maximum number entered. 
The program should: 
Keep taking input until the user enters a specific stop word (e.g., "stop"). 
Ignore the stop word in calculations. 
Display the maximum value after the input process ends. 
"""""

max_num = None
while True:
    user_input = input("Enter your number or 'Stop' to end")
    if user_input.lower() == "stop":
        break
    number = int(user_input)
    if max_num is None or number>max_num :
        max_num = number
print("The maximum number  is:", max_num)
```

```python
"""""
Problem 8:  
Write a function that takes any number of numbers and returns "All Even" if all numbers are even, "All Odd" if all are odd, and "Mixed" otherwise.
 """""
def check_numbers(*numbers):
    all_even = all(num % 2 == 0 for num in numbers)
    all_odd = all(num % 2 != 0 for num in numbers)

    if all_even:
        return "All Even"
    elif all_odd:
        return "All Odd"
    else:
        return "Mixed"


numbers = [int(x) for x in input("Enter numbers separated by spaces: ").split()]
result = check_numbers(*numbers)
print(result)
 
```

```python
"""""
Problem 9:  
Write a program to analyze a student’s scores and info. 
Keep asking the user for subject and score until "stop" is entered. 
Create functions: 
highestScore(**scores) → returns the subject with the highest score 
averageScore(*scores) → returns the average score 
gradeCategory(avg) → returns "Excellent", "Good", "Pass", or "Fail" 
studentInfo(*scores, **info) → prints student name and average 
After input ends, print: Highest score subject, Average score, Grade category, Student info 
"""""
scores = {}

while True:
    subject = input("Enter subject or 'stop' to end: ")
    if subject.lower() == "stop":
        break
    score_value = float(input(f"Enter score for {subject}: "))
    scores[subject] = score_value


def highest_score(**scores):
    return max(scores, key=scores.get)


def average_score(*scores):
    return sum(scores) / len(scores) if scores else 0


def grade_category(avg):
    if avg >= 90:
        return "Excellent"
    elif avg >= 75:
        return "Good"
    elif avg >= 60:
        return "Pass"
    else:
        return "Fail"


def studentInfo(*scores, **info):
    avg = average_score(*scores)
    print(f"Student Name: {info.get('name', 'Unknown')}")
    print(f"Average Score: {round(avg, 2)}")


student_name = input("Enter student name: ")

if scores:
    highest_subject = highest_score(**scores)
else:
    highest_subject = "No subjects entered"

avg_score = average_score(*scores.values())
category = grade_category(avg_score)

print("Highest Score Subject:", highest_subject)
print("Average Score:", round(avg_score, 2))
print("Grade Category:", category)

studentInfo(*scores.values(), name=student_name)
```

```python
"""""
Problem 10:  
Write a function called password_validator that takes a password string and returns "Strong" if it contains at least 8 characters, has at least one digit, and at least one uppercase letter. Otherwise, return "Weak". 

"""""
def password_validat(password):
    if (len(password) >= 8 and
        any(char.isdigit() for char in password) and
        any(char.isupper() for char in password)):
        return "Strong"
    else:
        return "Weak"
    
password = input("Enter your password: ")
print(password_validat(password))
```
