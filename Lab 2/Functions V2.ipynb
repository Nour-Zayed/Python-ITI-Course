```python
"""
Problem 1:  

Write a function that asks the user for a list of numbers separated by spaces, 
adds 10 to each number without exceeding 100, and returns the final list using map() and lambda. 
"""
def Add_Ten():
    user_input = input("Enter a list of numbers separated by spaces: ")
    Number = list(map(int,user_input.split()))
    result = list(map(lambda X: X + 10 if X + 10 <= 100 else X, Number))
    return result
print(Add_Ten())
     


```

```python
"""
Problem 2:  
Given a list of names, use map() to create a list of email addresses by adding "@company.com" to each name (lowercase). 
Input: ["Ahmed", "Sara", "Omar"] 
Output: ["ahmed@company.com", "sara@company.com", "omar@company.com"] 
"""
def Emails_Creation():

    names = ["Nour", "Mohamed", "Lina"]
    emails = list(map(lambda name: name.lower() + "@company.com", names))
    return emails
print(Emails_Creation())
```

```python
"""
Problem 3:  
Write a one-line function that reads a list of numbers as strings from the user,
 converts them to integers, doubles each value, converts the results back to strings,
   and returns the final list using map() only (no loops). 
"""
def Double_Numbers():
    return list(map(lambda X: str(int(X)*2), input("Enter a list of numbers: ").split()))
print(Double_Numbers())

```

```python
"""
Problem 4:  
Write a function that takes two lists from the user, asks for an operation (add, mul, str), applies it to corresponding elements using map() and lambda, handles different-length lists correctly, and uses no loops. 
 Example (Numbers) first list: 1 2 3 4, second list: 10 20    Output [11, 22] 
 Example (Strings)  first list: Py Data AI, second list: thon Science ML, Output ["Python", "DataScience", "AIML"] 
 """
def operate_Lists():
    list1 = input("Enter  first list (numbers or strings)  ").split()
    list2 = input("Enter  second list (numbers or strings)  ").split()
    operation = input("Enter operation (add, mul, str): ")

    if operation == "add":
        result = list(map(lambda x, y: int(x) + int(y), list1, list2))
    elif operation == "mul":
        result = list(map(lambda x, y: int(x) * int(y), list1, list2))
    elif operation == "str":
        result = list(map(lambda x, y: x + y, list1, list2))
    else:
        return "Invalid operation"

    return result
print(operate_Lists())
```

```python
"""
Problem 5:  
Given a list of words, keep only words that have more than 5 characters and contain the letter 'a'. 
Input: ["python", "programming", "code", "banana", "app"] 
Output: ["programming", "banana"] 
"""
def words_Filter():
    words = ["python", "programming", "code", "banana", "app"]
    filtered_words = list(filter(lambda word: len(word) > 5 and 'a' in word, words))
    return filtered_words
print(words_Filter())
```

```python
"""
Problem 6:  
Given a list of email addresses, use filter() to keep only valid emails (must contain @ and end with .com or .org). 
Input: ["test@email.com", "invalid.email", "user@site.org", "bad@"] 
Output: ["test@email.com", "user@site.org"] 
"""
def validate_Emails():
    emails = ["test@email.com", "invalid.email", "user@site.org", "bad@"]
    valid_Emails = list(filter(lambda email: '@' in email and (email.endswith('.com') or email.endswith('.org')), emails))
    return valid_Emails
print(validate_Emails())
```

```python
"""
Problem 7:  
Given a list of numbers, use reduce() to calculate the product of all numbers. 
Input: [2, 3, 4, 5]  -  Output: 120 
"""
from functools import reduce
def product_of_Numbers():
    numbers = [2, 3, 4, 5]
    product = reduce(lambda x, y: x * y, numbers)
    return product
print(product_of_Numbers())
```

```python
"""
Problem 8:  
Given a list of numbers entered by the user, select only the even numbers, then square them, and finally calculate the sum of all squared numbers. 
 Hint: use filter , Map and reduce . No loops allowed. 
 """
from functools import reduce
def squard_Even_Sum():
    user_input = input("Enter  Your  numbers  ")
    numbers = list(map(int, user_input.split()))
    even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
    squared_numbers = list(map(lambda x: x ** 2, even_numbers))
    total_sum = reduce(lambda x, y: x + y, squared_numbers, 0)
    return total_sum
print(squard_Even_Sum())
```

```python
"""
 Problem 9:  
Given a list of prices with currency symbols ("$50", "$120", "$30"), remove the $ and convert it to integers, keep prices above $40, then calculate the average. 
"""
def AVG_High_Prices():
    prices = ["$50", "$120", "$30", "$80", "$25"]
    numeric_prices = list(map(lambda price: int(price.replace('$', '')), prices))
    high_prices = list(filter(lambda price: price > 40, numeric_prices))
    average_price = sum(high_prices) / len(high_prices) if high_prices else 0
    return average_price
print(AVG_High_Prices())
```

```python
"""
Bonus 1:  
Given a list of numbers, use map() to create a new list where each number is replaced with "Even" or "Odd". 
Input: [1, 2, 3, 4, 5] 
Output: ["Odd", "Even", "Odd", "Even", "Odd"]
 """
def even_Odd_Labels():
    numbers = [1, 2, 3, 4, 5]
    labels = list(map(lambda x: "Even" if x % 2 == 0 else "Odd", numbers))
    return labels
print(even_Odd_Labels())
```

```python
"""
Bonus 2:  
Write a function that takes a list of numbers and a list of operations (functions). Apply all operations in sequence to each number using reduce() on the operations list and map() on the numbers. 
Input: numbers = [5, 10], operations = [lambda x: x*2, lambda x: x+10, lambda x: x**2] 
Output: [400, 900] (each number: ×2 → +10 → ²) 
"""

from functools import reduce
def apply_Operations(numbers, operations):
    return list(map(lambda x: reduce(lambda acc, func: func(acc), operations, x), numbers)) 
numbers = [5, 10]   
operations = [lambda x: x*2, lambda x: x+10, lambda x: x**2]
print(apply_Operations(numbers, operations))
```

```python
"""
Bonus 3:  
Create a function called custom_sort that takes a list and a comparison function (passed as an argument). The comparison function should determine the sorting order. Implement this without using Python's built-in sorted() or .sort(). 
Hint: Use bubble sort or selection sort with your custom comparison function. 
"""
def custom_sort(lst, cmp):
    lst = lst[:]   

    for i in range(len(lst)):
        for j in range(len(lst) - 1):
            if cmp(lst[j], lst[j + 1]):
                lst[j], lst[j + 1] = lst[j + 1], lst[j]

    return lst


nums = [5, 2, 9, 1]

print(custom_sort(nums, lambda a, b: a > b))

print(custom_sort(nums, lambda a, b: a < b))

words = ["AI", "Python", "ML", "Data"]

print(custom_sort(words, lambda a, b: len(a) > len(b)))


```
