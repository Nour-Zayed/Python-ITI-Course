```python

"""Problem 1:  

Write a function that creates a text file and writes a list of names to it, each name on Write a Python program with two functions: 

Write & Read Names 

Create a text file and write a list of names, one per line. 

Read the file and return the names as a list. 

Example Input: ["Ahmed", "Sara", "Omar"] 
Expected Output: ["Ahmed", "Sara", "Omar"] 

Word Count with Exception Handling 

Read a text file and count how many times each word appears. 

Return a dictionary of word counts. 

Handle FileNotFoundError if the file does not exist. 

Example File Content: "hello world hello python" 
Expected Output: {"hello": 2, "world": 1, "python": 1} 
"""

names = ["Ahmed", "Nour", "Mohamed"]
with open("D:/Python ITI/Lab 6/names.txt", "w", encoding="utf-8") as f:
    f.writelines(name + "\n" for name in names)

with open("D:/Python ITI/Lab 6/names.txt", "r", encoding="utf-8") as f:
    read_names = [line.strip() for line in f]
print("Names:", read_names)

with open("D:/Python ITI/Lab 6/example.txt", "w", encoding="utf-8") as f:
    f.write("hello world hello python")

try:
    with open("D:/Python ITI/Lab 6/example.txt", "r", encoding="utf-8") as f:
        text = f.read()
        words = text.split()
        counts = {}
        for word in words:
            counts[word] = counts.get(word, 0) + 1
    print("Word counts:", counts)
except FileNotFoundError:
    print("example.txt not found!")



```

```python
"""
Problem 2:  
Write a function that reads a JSON file, adds a new key-value pair to the data, and writes it back to the file. Handle FileNotFoundError and json.JSONDecodeError. 
Original JSON: {"name": "Ahmed"} 
Add: "age": 25 
Result: {"name": "Ahmed", "age": 25} 
"""

import json
import os

filepath = "D:/Python ITI/Lab 6/test.json"

if not os.path.exists(filepath):
    data = {"name": "Ahmed"}
else:
    with open(filepath, "r", encoding="utf-8") as file:
        try:
            data = json.load(file)
        except json.JSONDecodeError:
            print("Warning: Invalid JSON. Starting fresh.")
            data = {}

print("Original JSON:", data)

data["age"] = 25

with open(filepath, "w", encoding="utf-8") as file:
    json.dump(data, file, indent=4)

print("Updated JSON:", data)




```
