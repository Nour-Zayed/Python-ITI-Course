```python
"""
Problem 1:  
Write a function that calculates the distance between two points (x1, y1) and (x2, y2) using the formula: √((x2-x1)² + (y2-y1)²). Use math.sqrt() and math.pow(). 
"""

import math

def distance(x1, y1, x2, y2):
    return math.sqrt(math.pow(x2 - x1, 2) + math.pow(y2 - y1, 2))
print(distance(1, 2, 4, 6))  
```

```python
"""
Problem 2:  
Write a function that takes a birthdate (year, month, day) and calculates the person's age in years. Use the datetime module. 
"""

from datetime import datetime
def calculate_age(year, month, day):
    today = datetime.today()
    birthdate = datetime(year, month, day)
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age
print(calculate_age(2002, 11, 1))
```

```python
"""
Problem 3:  
Write a function that takes two dates and calculates the number of days between them. Use datetime module. 
"""
from datetime import datetime
def Cal_days(date1, date2):
    d1 = datetime.strptime(date1, "%Y-%m-%d")
    d2 = datetime.strptime(date2, "%Y-%m-%d")
    return abs((d2 - d1).days)
print(Cal_days("2025-01-01", "2025-12-31"))
```

```python
"""
Bonus: 

Problem 1:  
Write a function that calculates the number of working days between two dates (excluding weekends). 
Input: "2024-01-01", "2024-01-15" 
Output: 11 working days 
"""
from datetime import datetime, timedelta
def work_days(date1, date2):
    d1 = datetime.strptime(date1, "%Y-%m-%d")
    d2 = datetime.strptime(date2, "%Y-%m-%d")
    delta = timedelta(days=1)
    working_days_count = 0
    
    while d1 <= d2:
        if d1.weekday() < 5:  
            working_days_count += 1
        d1 += delta
        
    return working_days_count   
print(work_days("2024-01-01", "2024-01-15"))
```
