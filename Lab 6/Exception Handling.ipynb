```python
"""
Problem 1:  
Write a function that takes a list and an index, then returns the element at that index. Handle both IndexError (invalid index) and TypeError (if input is not a list). 
Input: [1, 2, 3], 5 → Output: "Index out of range" 
Input: "hello", 2 → Output: "Invalid input type" 

"""

def exp_handling(ls,index):
    try:
        return ls[index]
    except TypeError:
        return "Invalid input type"
    except IndexError:
        return "Index out of range" 

print(exp_handling([1, 2, 3], 5))

print(exp_handling("hello",2 ))


```

```python
"""Problem 2:  
Write a function that takes a dictionary and a key, then returns the value. Use try-except to handle KeyError, and in the finally block, print "Search completed" regardless of success or failure. 
Input: {"name": "Ahmed", "age": 25}, "name" → Output: "Ahmed" + "Search completed" 
Input: {"name": "Ahmed"}, "age" → Output: "Key not found" + "Search completed" 
"""
def dict_handling(dct,key):
    try:
        return dct[key]
    except KeyError:
        return "Key not found"
    finally:
        print("Search completed")

print(dict_handling({"name": "Ahmed", "age": 25}, "name"))
print(dict_handling({"name": "Ahmed"}, "age"))
```

```python
"""
Problem 3:  
Create a custom exception InvalidEmailError with an error code. Write a function that validates an email (must contain @ and end with .com or .org). Raise the custom exception with error code 1001 if invalid. 
Input: "user@example.com" → Output: "Valid email" 
Input: "userexample.com" → Output: "[Error 1001] Invalid email format"
 """
class InvalidEmailError(Exception):
    def __init__(self, message, error_code):
        self.message = message
        self.error_code = error_code
        super().__init__(self.message)

def validate_email(email):
    try:
        if "@" not in email or not email.endswith((".com", ".org")):
            raise InvalidEmailError("Invalid email format", 1001)
        return "Valid email"
    except InvalidEmailError as e:
        return f"[Error {e.error_code}] {e}"

print(validate_email("user@example.com"))
print(validate_email("userexample.com"))
```
