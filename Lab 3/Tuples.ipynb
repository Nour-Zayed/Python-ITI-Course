```python
"""Problem 1:  

Write a function that takes a tuple of strings and returns the longest string. If there are multiple strings with the same maximum length, return a tuple containing all of them. 
Inputs: ("hi", "hello", "hey", "python") | ("hello", "world", "python") 
Outputs: 	"python"  | ("hello", "world") (both are 5 chars) 
"""
from functools import reduce
tip1= ("hi", "hello", "hey", "python") 
tip2 = ("hello", "world", "python")
max1 = reduce(lambda a, b: a if len(a) > len(b) else b, tip1)
max2 = reduce(lambda a, b: a if len(a) > len(b) else b, tip2)

t1 = tuple(filter(lambda i : len(i) == len(max1),tip1))
t2 = tuple(filter(lambda i : len(i) == len(max2),tip2))

print(t1)
print(t2)
```

```python
"""Problem 2:  

Write a function that takes a tuple and an element, then returns the number of times that element appears in the tuple and all the positions (indices) where it appears. 
Input: (1, 2, 3, 2, 4, 2, 5), 2 
Output: "Count: 3, Positions: (1, 3, 5)" 
"""
t = (1, 2, 3, 2, 4, 2, 5)
def tu_element(t,num):
    pos = ()
    for i in range(len(t)):
        if t[i] == num:
            pos += (i,)
    return pos 
pos = tu_element(t,2)
print(t)
print('Count:',len(pos),'Positions:',pos)

```

```python
"""Problem 3:  

Write a function that uses tuple unpacking with asterisk (*) to take a tuple of any length and return the first element, last element, and the sum of all middle elements. 
Input: (10, 20, 30, 40, 50) 
Output: "First: 10, Last: 50, Middle sum: 90" 
"""
tip = (10, 20, 30, 40, 50)
def tup_unpacking(tip):
    f,*mid,l = tip
    return f,sum(mid),l
res = tup_unpacking(tip)
print('First =',res[0],", Last =",res[2],', Sum Of Middle Elements =',res[1])
```

```python
"""Problem 4:  

Write a function that takes a tuple of words and returns a new tuple where each word is reversed, but the order of words remains the same. 
Input: ("hello", "world", "python") 
Output: ("olleh", "dlrow", "nohtyp") 
"""
t= ("hello", "world", "python")
print(t)
def tuple_reverse(t):
    t2 = ()
    for i in t:
        t2 += (i[::-1],)
    return t2
print(tuple_reverse(t))
```
