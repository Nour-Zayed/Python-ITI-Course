```python
"""Problem 1
Write a function that takes two lists of equal length and returns a dictionary where elements from the first list are keys and elements from the second list are values. 
Input: ["name", "age", "city"], ["Ali", 25, "Cairo"] 
Output: {"name": "Ali", "age": 25, "city": "Cairo"} 
"""

lst1= ["name", "age", "city"]
lst2 = ["Ali", 25, "Cairo"]
def list_Dic(ls1,ls2):
    dic = {}
    for i in range(len(lst1)):
        dic[lst1[i]] = lst2[i]
    return dic
print(list_Dic(lst1,lst2))
```

```python
"""
Problem 2:  
Write a function that takes a dictionary of product names and prices, then returns a new dictionary containing only products that cost more than 50, with prices increased by 10%. 
Input: {"apple": 30, "banana": 60, "orange": 80} 

Output: {"banana": 66.0, "orange": 88.0} 
"""
dic =  {"apple": 30, "banana": 60, "orange": 80}
def price_inc(d):
    d2={}
    for M,N in d.items():
        if N > 50:
            d2[M] = N *1.1
    return d2
print(price_inc(dic))
```

```python
"""
Problem 3:  
Write a function that takes a dictionary where keys are student names and values are lists of grades. Return a new dictionary with the same keys, but filter out any grades below 60, then calculate the average of remaining grades. 
Input: {"Ahmed": [70, 50, 80, 55], "Sara": [90, 85, 95]} 
Output: {"Ahmed": 75.0, "Sara": 90.0} 
"""
dir = {"Ahmed": [70, 50, 80, 55], "Sara": [90, 85, 95]}
def avg_grade(d):
    dic = {}
    for M,N in d.items():
        ls = list(filter(lambda i: i >=60 ,N))
        if ls:  
            dic[M] = sum(ls)/ len(ls) 
        else:
            dic[M] = 0  
    return dic
print(avg_grade(dir))
```

```python
"""Problem 5:  

Write a function that takes a dictionary where keys are names and values are ages, then groups people into 
categories: "child" (0-12), "teen" (13-19), "adult" (20-59), "senior" (60+). Return a dictionary with these categories as keys and lists of names as values. 
Input: {"Ahmed": 25, "Sara": 15, "Omar": 65, "Ali": 10} 
Output: {"child": ["Ali"], "teen": ["Sara"], "adult": ["Ahmed"], "senior": ["Omar"]} 
Output: {"Ahmed": ["Math", "English"], "Sara": ["Math", "Science"], "Omar": ["Science"]} 
"""
dir = {"Ahmed": 25, "Sara": 15, "Omar": 65, "Ali": 10}
def categ(d):
    dic = {}
    child =[]
    teen =[]
    adult=[]
    senior = []
    for M,N in d.items():
        if N < 12:
            child.append(M)
        elif N < 19:
            teen.append(M)
        elif N < 60:
            adult.append(M)
        else:
            senior.append(M)
    dic['child'] = child ; dic['teen'] = teen ; dic['adult'] = adult ; dic['senior'] = senior
    return dic
print(categ(dir))


```
