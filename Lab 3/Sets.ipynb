```python
"""
Problem 1:  
Write a function that takes two sets and returns three sets: elements only in first set, elements only in second set, and elements in both (union, difference operations). 
Input: {1, 2, 3}, {2, 3, 4} 
Output: Only in first: {1}, Only in second: {4}, In both: {2, 3} 
"""
st1 =  {1, 2, 3}
st2 = {2, 3, 4}
print('Only in 1',st1.difference(st2))
print('Only in 2',st2.difference(st1))
print('Both',st1.intersection(st2))
```

```python
"""Problem 2:  

Write a function that takes a list of words and returns a set of words that are palindromes  
Input: ["level", "hello", "radar", "world", "civic"] 
Output: {"level", "radar", "civic"} 
"""
ls = ["level", "hello", "radar", "world", "civic"]
def palindromes(ls):
    result = set()
    for i in ls:
        if i == i[::-1]:
            result.add(i)
    return result
            
palindromes(ls)
```

```python
"""Problem 3:  

Create a frozenset from a list of tuples representing coordinates. Write a function that checks if a given coordinate exists in the frozenset. 
Frozenset: {(1, 2), (3, 4), (5, 6)} 
Check: (3, 4) → Output: True 
Check: (7, 8) → Output: False 
"""
tub = frozenset([(1, 2), (3, 4), (5, 6)])

def check(frset, tub):
    return tub in frset
##  frozenset ----->>> immutable set 
print(check(tub, (3, 4)))  
print(check(tub, (7, 8)))  
```
