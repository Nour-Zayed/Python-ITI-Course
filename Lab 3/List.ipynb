```python
"""
Problem 1:  
Fill an array of 5 elements from the user, Sort it in descending and ascending orders then display the output. 
 """
arr = []
for i in range(5):
    num = int(input(f"Enter element {i+1}: "))
    arr.append(num)

asc = sorted(arr)
desc = sorted(arr, reverse=True)

print("Array in ascending order:", asc)
print("Array in descending order:", desc)
```

```python
"""
Problem 2:  
Write a function that takes a list of numbers and returns a new list containing only the numbers that are greater than the average of all numbers in the original list. 
Input: [10, 20, 30, 40, 50] 
Output: [40, 50] (average is 30) 

 """
def above_average(numbers):
    avg = sum(numbers) / len(numbers)
    result = [num for num in numbers if num > avg]
    return result

nums = [10, 20, 30, 40, 50]
print("Average:", above_average(nums))
```

```python
"""
Problem 3:  
Write a function that takes a list of numbers and removes all duplicates while keeping the original order. Return the modified list and the count of removed duplicates. 
Input: [1, 2, 3, 2, 4, 1, 5] 
Output: [1, 2, 3, 4, 5], "Removed 2 duplicates" 
"""
def remove_duplicates(lst):
    seen = set()
    result = []
    removed_count = 0

    for num in lst:
        if num not in seen:
            result.append(num)
            seen.add(num)
        else:
            removed_count += 1

    return result, removed_count

nums = [1, 2, 3, 2, 4, 1, 5]
new_list, removed = remove_duplicates(nums)
print("Modified list:", new_list)
print(f"Removed {removed} duplicates")
```

```python
"""
Problem 4:
 Write a function that takes a list of mixed data types (numbers, strings, booleans) 
 and returns three separate lists: one for numbers, one for strings, and one for booleans. 
 Input: 
 Output: [1, 3.5, 42], ["hello", "world"], [True, False]
"""

def separate_types(lst):
    numbers = []
    strings = []
    booleans = []

    for item in lst:
        if isinstance(item, bool):
            booleans.append(item)
        elif isinstance(item, (int, float)):
            numbers.append(item)
        elif isinstance(item, str):
            strings.append(item)

    return numbers, strings, booleans

mixed_list = [1, "hello", True, 3.5, "world", False, 42]
nums, strs, bools = separate_types(mixed_list)
print("Numbers:", nums)
print("Strings:", strs)
print("Booleans:", bools)
```

```python
"""
Problem 5: 
Write a function that takes a 2D list (matrix) and returns the sum of all elements on the main diagonal
( top-left to bottom-right). 
Input: [1, 2, 3], [4, 5, 6], [7, 8, 9]] Output: 15 (1 + 5 + 9)
"""
def diagonal_sum(matrix):
    total = 0
    for i in range(len(matrix)):
        total += matrix[i][i]  
    return total

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("Sum of main diagonal:", diagonal_sum(matrix))  

```

```python
"""
Problem 6: 
Write a function that takes a list of numbers, finds the two smallest numbers, removes them from the list, and returns their sum along with the modified list. 
Input: [5, 2, 8, 1, 9] 
Output: Sum: 3, List: [5, 8, 9]
"""

def sum_two_smallest(lst):
    if len(lst) < 2:
        return "List must have at least 2 numbers."

    lst_copy = lst.copy()
    first_min = min(lst_copy)
    lst_copy.remove(first_min)
    second_min = min(lst_copy)
    lst_copy.remove(second_min)

    sum_min = first_min + second_min
    return sum_min, lst_copy

numbers = [5, 2, 8, 1, 9]
sum_min, modified_list = sum_two_smallest(numbers)
print("Sum:", sum_min)
print("Modified list:", modified_list)
```
