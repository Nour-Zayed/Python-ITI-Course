```python

#Problem 1: 
#Input: two numbers a and b and an integer operator
#If operator == 1, add a + b
#If operator == 2, subtract b from a
#If operator == 3, multiply a * b
#Otherwise, print "Invalid Input"
#Provide two solutions
a = int(input())
b = int(input())
operator = int(input())

if operator == 1:
    print(a + b)
elif operator == 2:
    print(b - a)
elif operator == 3:
    print(a * b)
else:
    print("Invalid Input")
    
    
a = int(input())
b = int(input())
operator = int(input())

match operator:
    case 1:
        print(a + b)
    case 2:
        print(a - b)
    case 3:
        print(a * b)
    case _:
        print("Invalid Input")

```

```python
#Problem 2: 
#Input: two integers n and m (m != 0)
#Output: the number closest to n divisible by m
#If there are two candidates, choose the one with the maximum absolute value

n,m  = map (int, input().split())
lower = (n // m) * m
upper = lower + m
if abs(lower) > abs(upper):
    print(lower)
else:
    print(upper)
```

```python

#Problem 3: 
#Input: an integer n
#Output: True if n is prime, False otherwise
n = int(input())
if n <= 1:
    print(False)
else:
    is_prime = True
    for i in range (2,int(n**.5)+1):
        if n % i == 0:
            is_prime = False
            break

    print(is_prime)
```

```python
#Problem 4: Course Admission Eligibility 
"""""
Create a program that asks the user for: 

Age, Number of completed projects, Knows JavaScript (Yes/No), Score in previous course 

Eligibility Rules: 

Age >= 20 

Completed projects >= 2 OR knows JavaScript 

Score >= 70 

Requirements: 

Classify experience using ternary: "Advanced" if projects >= 5, no "Beginner" 

Print reasons if not eligible 

If eligible, print: "Eligible for the course. Experience level: Advanced/Beginner" """""

Age = int(input("Enter your age: "))
Projects = int(input("Enter number of completed projects: "))
Knows_JavaScript = input("Do you know JavaScript? (Yes/No): ")
Score = int(input("Enter your score in previous course: "))

eligible = True
reasons = []

if Age < 20:
    eligible = False
    reasons.append("Age is less than 20.")
if Projects < 2 and Knows_JavaScript == "No":
    eligible = False
    reasons.append("Not enough projects and does not know JavaScript.")
if Score < 70:
    eligible = False
    reasons.append("Score is less than 70.")

if eligible:
    experience_level = "Advanced" if Projects >= 5 else "Beginner"
    print(f"Eligible for the course. Experience level: {experience_level}")
else:
    print("Not eligible for the course.")
    for reason in reasons:
        print(f"- {reason}")
```

```python
""""
Problem 5:  
Ask the user to enter their age. 
Prints whether the age input is truthy. 
Classifies the user based on age using multiple conditions: 
18 or older → "Adult" 
Between 13 and 17 → "Teenager" 
12 or younger → "Child" 
Empty or invalid input → "Invalid age" 
"""""
Age_input = input("Enter your age: ")
print("Truthy input:", bool(Age_input))
if not Age_input.isdigit():
    print("Invalid age")
else:
    Age = int(Age_input)
    if Age >= 18:
        print("Adult")
    elif 13 <= Age < 18:
        print("Teenager")
    else:
        print("Child")
```

```python
""""
Problem 6:  

Ask the user to input a password. 
Check if the password is truthy. 
Print different messages: 
Password length ≥ 8 → "Strong password" 
Password length < 8 and non-empty → "Weak password" 
Empty input → "No password entered" 
Bonus: also check if the password contains both letters and digits. 
"""""

password = input("Enter password: ")

if not password:
    print("No password entered")
else:
    has_letter = False
    has_digit = False

    for ch in password:
        if ch.isalpha():
            has_letter = True
        elif ch.isdigit():
            has_digit = True

    if len(password) >= 8:
        if has_letter and has_digit:
            print("Strong password (contains letters and digits)")
        else:
            print("Strong password (but should contain both letters and digits)")
    else:
        print("Weak password")

```
