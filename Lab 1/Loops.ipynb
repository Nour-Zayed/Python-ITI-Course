```python
"""""
Problem 1:  
Input: a string s 
Output: print all characters at even indices 
Example: 
Input: "DoctorPhenomenal" → Output: "DcoPeoea" 
"""""
s = input("Enter your string: ")
result = ""
for i in range(0, len(s), 2):
    result += s[i]
print("Characters  even indices:", result)    

```

```python
""""Problem 4:  
Write a program which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake, print an error message and skip to the next number. 
"""""
total = 0
count = 0
while True:
    user_input = input("Enter A num or 'done' to finish: ")
    if user_input.lower() == 'done':
        break
    try:
        number = float(user_input)
        total += number
        count += 1
    except ValueError:
        print("Invalid input. Please enter a number.")

if count > 0:
    average = total / count
    print("Total:", total)
    print("Count:", count)
    print("Average:", average)
else:
    print("No valid numbers were entered.")
```

```python
""""
Problem 5:  
Input: an integer n 
Output: factorial of n (1*2*3*...*n) 
"""""
n = int(input("Enter a positive num : "))
fact  = 1
for i in range(1,n+1):
    fact*=i
print(fact)    
```

```python
"""""
Problem 6:  
Input: an integer n 
Output: the nth Fibonacci number (F(0)=0, F(1)=1, F(n)=F(n-1)+F(n-2)) 
"""""
n = int(input("Enteryour num "))
if n == 0:
    print(0)
elif n == 1:
    print(1)
else:
    a,b = 0,1
    for _ in range(2,n+1):
        a,b = b,a+b
    print(b)
```

```python
""""" Problem 7:  

Ask the user for an integer n. 

Print the following pattern: 
      *
     ***
    *****
   *******
  *********
 ***********
 """""
n = int(input("Enter  number of row "))
for i in range(n):
    print(' ' * (n - i - 1) + '*' * (2 * i + 1))
```

```python
"""""
Problem 8:  
Ask the user for an integer n. 
Print numbers from 1 to n. 
For multiples of 3, print "Fizz" instead of the number. 
For multiples of 5, print "Buzz". 
For multiples of both 3 and 5, print "FizzBuzz". 
Extend: store the output in a list and print the final list at the end. 
"""""
n = int(input("Enter  num"))
results = []
for i in range (1,n+1):
    if i % 3 == 0 and i % 5 == 0:
        results.append("FizzBuzz")
    elif i % 3 == 0:
        results.append("Fizz")
    elif i % 5 == 0:
        results.append("Buzz")
    else:
        results.append(str(i))
print("\n".join(results))
```

```python
"""""
Bonus: 
Problem 1:  
Ask the user to enter a positive integer. 
Count how many times each digit (0-9) appears in the number using a loop. 
Print the counts in a dictionary format. 
"""""

num = input("Enter num  ")
counts ={}
for digit in num:
    counts[digit] = counts.get(digit,0) + 1
print("Digit counts:", counts)    
```

```python
"""""
Problem 2:  
Ask the user to input a square matrix (nested lists). 
Using nested loops, compute the sum of the main diagonal and sum of the secondary dia	gonal. 
Print both sums. 
"""""

n = int(input("Enter the size of matrix: "))
matrix = []
for i in range(n):
    row = list(map(int, input(f"Enter row {i+1} (space-separated): ").split()))
    matrix.append(row)

main_diagonal_sum = 0
secondary_diagonal_sum = 0
for i in range(n):
    
    main_diagonal_sum += matrix[i][i]
    secondary_diagonal_sum += matrix[i][n-i-1]

print("Main diagonal :", main_diagonal_sum)
print("Secondary diagonal :", secondary_diagonal_sum)
```
